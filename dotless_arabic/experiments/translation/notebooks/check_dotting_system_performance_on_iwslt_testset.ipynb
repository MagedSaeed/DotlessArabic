```python
from tqdm.auto import tqdm
from dotless_arabic.processing import undot
from dotless_arabic.experiments.translation.src.processing import process_ar
from dotless_arabic.datasets.iwslt2017.collect import collect_parallel_test_dataset_for_translation
from dotless_arabic.experiments.dots_retrieval.src.utils import add_dots_to_undotted_text, calculate_text_metrics_from_text_inputs
```

```python
dataset = collect_parallel_test_dataset_for_translation()['ar']
dataset
```

```python
processed_dataset = list(map(process_ar,dataset))
processed_dataset
```

```python
undotted_processed_dataset = list(map(undot,processed_dataset))
undotted_processed_dataset
```

```python
artificially_dotted_dataset = list(map(add_dots_to_undotted_text,tqdm(undotted_processed_dataset)))
artificially_dotted_dataset
```

```python
wers = list()
cers = list()
for predicted_text,true_text, in zip(artificially_dotted_dataset,processed_dataset):
    wer,cer = calculate_text_metrics_from_text_inputs(
        predicted_text=predicted_text,
        true_text=true_text,
    )
    wers.append(wer)
    cers.append(cer)
sum(wers)/len(wers),sum(cers)/len(cers)
```
