```python
import pandas as pd
import matplotlib.pyplot as plt
from dotless_arabic.datasets.poems.collect import collect_dataset_for_meter_classification
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_classes_count(dataset):
    class_counts = {}
    
    # Count occurrences of each class
    for value in dataset.values():
        if value in class_counts:
            class_counts[value] += 1
        else:
            class_counts[value] = 1
    
    # Convert dictionary to DataFrame for Seaborn
    class_data = {'Class': list(class_counts.keys()), 'Count': list(class_counts.values())}
    class_df = pd.DataFrame(class_data)
    
    # Plotting
    plt.figure(figsize=(8, 6))
    sns.barplot(x='Class', y='Count', data=class_df, color='brown')
    plt.xlabel('Classes')
    plt.ylabel('Count')
    plt.title('Count of Instances per Class')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
d = collect_dataset_for_meter_classification()
```

```python
plot_classes_count(dataset=d)
```
